Calculate total sales from units and price per unit using matrix multiplication

```python
price_per_unit = np.array([1000,400,1200])
units = np.array([[30,40,50],[5,10,15],[2,5,7]])
```

```python
price_per_unit*units
```


In above case numpy is using broadcasting so it expands price_per_unit array from 1 row, 3 columns to 3 row and 3 columns. Correct way to do matrix multiplication is to use dot product as shown below

```python
np.dot(price_per_unit,units)
```

```python
import numpy as np

data_usd = np.array([
    [200, 220, 250],  # Reliance Industries
    [68, 79, 105],    # Infosys
    [110, 140, 180],  # TCS
    [80, 85, 90]      # Wipro
])
```

```python
usd_to_inr = 75
```

```python
data_inr = data_usd * usd_to_inr
```

```python
import pandas as pd

companies = ["Reliance Industries", "Infosys", "TCS", "Wipro"]
years = ["FY 2018", "FY 2019", "FY 2020"]

df_inr = pd.DataFrame(data_inr, index=companies, columns=years)
print(df_inr)
```

```python
import numpy as np
import pandas as pd

# Step 1: Units sold for each flower (rows) across months (columns)
units_sold = np.array([
    [50, 60, 25],  # Roses
    [10, 13, 5],   # Lotus
    [40, 70, 52]   # Sunflower
])

# Step 2: Price per flower (in )
price_per_unit = np.array([20, 30, 15])  # Roses, Lotus, Sunflower

# Step 3: Calculate total sales per month
# Multiply prices (row vector)  units_sold (matrix)
total_sales = np.dot(price_per_unit.reshape(1, 3), units_sold).flatten()

# Step 4: Create a readable DataFrame
months = ["Jan", "Feb", "March"]
df_sales = pd.DataFrame([total_sales], columns=months, index=["Total Sales ()"])

# Step 5: Display result
print(df_sales)
```
