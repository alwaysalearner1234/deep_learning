Implement Mean Absolute Error

```python
def mae(y_predicted, y_true):
    total_error = 0
    for yp, yt in zip(y_predicted, y_true):
        total_error += abs(yp - yt)
    print("Total error is:",total_error)
    mae = total_error/len(y_predicted)
    print("Mean absolute error is:",mae)
    return mae
```

```python
import numpy as np

# Example actual and predicted values
y_true = np.array([10, 20, 30, 40])
y_predicted = np.array([12, 18, 33, 39])

# Define a function for Mean Absolute Error (MAE)
def mae(y_predicted, y_true):
    return np.mean(np.abs(y_predicted - y_true))

# Now call it
error = mae(y_predicted, y_true)
print("Mean Absolute Error:", error)
```

```python
def log_loss(y_true, y_predicted):
    y_predicted_new = [max(i,epsilon) for i in y_predicted]
    y_predicted_new = [min(i,1-epsilon) for i in y_predicted_new]
    y_predicted_new = np.array(y_predicted_new)
    return -np.mean(y_true*np.log(y_predicted_new)+(1-y_true)*np.log(1-y_predicted_new))
```

Implement mean squared error (or MSE) in two ways,

Without using numpy (i.e. using plain python)

With the use of numpy

```python
# Mean Squared Error without using NumPy
def mse_plain(y_true, y_pred):
    # ensure both lists have same length
    n = len(y_true)
    squared_errors = [(y_pred[i] - y_true[i]) ** 2 for i in range(n)]
    mse = sum(squared_errors) / n
    return mse

# Example
y_true = [1, 2, 3, 4]
y_pred = [1.2, 1.9, 3.3, 3.8]

print("MSE (plain Python):", mse_plain(y_true, y_pred))
```

```python
import numpy as np

# Mean Squared Error using NumPy
def mse_numpy(y_true, y_pred):
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    mse = np.mean((y_pred - y_true) ** 2)
    return mse

# Example
y_true = [1, 2, 3, 4]
y_pred = [1.2, 1.9, 3.3, 3.8]

print("MSE (NumPy):", mse_numpy(y_true, y_pred))
```
